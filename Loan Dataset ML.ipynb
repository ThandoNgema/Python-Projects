```python
import pandas as pd
import matplotlib as plt

loans_df = pd.read_csv(r"C:\Users\user\Desktop\2026\Data Analytics\Data Sets\loans\loans_full_schema.csv")
loans_df.head()
```

```python
loans_df.columns
```

```python
loans_df = loans_df[['emp_length','annual_income','debt_to_income','balance','paid_total','paid_principal','paid_interest','paid_late_fees','interest_rate','term','loan_amount']]

loans_df
```

```python
loans_df.isnull().any()
```

```python
loans_df = loans_df.dropna()
loans_df.isnull().any()
```

```python
loans_df.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

k = 95 #square of 9182 rounded to the nearest odd number

knn = KNeighborsClassifier(n_neighbors=k)
sc = StandardScaler()

x = loans_df[['emp_length','annual_income','debt_to_income','interest_rate','term']] #Other columns can be used for EDA
y = loans_df[['loan_amount']]

x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.25, random_state = 42)

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

knn.fit(x_train, y_train)
knn
```

```python
predict_loans = knn.predict(x_test)
acc_score = accuracy_score(y_test, predict_loans)
print(acc_score)
```

## Conclusion: KNN model is not a great fit, I am going to try Linear Regression

```python
#Already import the other libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#Already have our x and y, we do not need a standard scaler so I'll retrain the data so it becomes unfitted and untransformed


x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.25, random_state = 42)

lm = LinearRegression()

lm.fit(x_train, y_train)
lm
```

```python
predicted_loans = lm.predict(x_test)
linear_acc_score = r2_score(y_test, predicted_loans)
print(linear_acc_score)
```

## Linear Model is 26% accurate, cannot be adopted and may need more training data to make better and accurate predictions
