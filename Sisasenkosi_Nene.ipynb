### 1. Write a program to display all odd numbers between a given range. [5]

```python
start = int(input('Enter starting range'))
end = int(input('Enter stopping range'))

for i in range(start,end):
    if i % 2 == 1:
        print(i)
    else:
        continue

```

### 2. Write a program to find the total count of a substring in a given string ignoring the case. (e.g. find the total count of the substring "data" in the string "I am learning data analytics so that I can be a Data Analytics professional". [5]


```python
word = input('Enter a word to find the number of its occurrence').lower()
sentence = input('Enter a sentence').lower()

count = 0

for i in sentence.split(' '):
    if i == word:
        count += 1
    else:
        continue

print(count)
```

### 3. Write a program to store 10 random integers between 1 and 20 into a list. Display the list. [6]

```python
import random as rnd
random_nums = []
count = 1
while count <= 10:
    num = round((rnd.random()*100))
    if num >=1 and num <=20: 
        random_nums.append(num) 
        count += 1
    else:
        continue

print(random_nums)
    
```

### 4. Find the highest and lowest integers in the list. [4]

```python
min = random_nums[0]
max = random_nums[0]
for i in random_nums:
    if i >= max:
        max = i
    elif i <= min:
        min = i
print(random_nums)
print(min)
print(max)
```

### 12. Build a machine learning model that can predict whether or not a student is selected based on skills_score, projects_count, internships_done, communication_score, aptitude_score, coding_test_score, resume_score, extracurricular, college_tier, hackathons_participated, certifications_count, linkedin_activity_score, github_score, soft_skills_score, interview_score, consistency_score, and placement_training. Training and testing sets should be split on a ratio of 75:25. Evaluate the performance of the model using an appropriate evaluation metric. [15]

```python
#Importing all the libraries I may or may not use

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, accuracy_score
```

```python
#Selecting my features
x = new_internship_df[['skills_score','projects_count','internships_done','communication_score','aptitude_score','coding_test_score','resume_score','extracurricular','college_tier','hackathons_participated','certifications_count','linkedin_activity_score','github_score','soft_skills_score','interview_score','consistency_score','placement_training']]
y = new_internship_df[['selected']]

#Split the training and testing data

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)

#Standardizing the data

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

#Train the model
k = 9
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(x_train, y_train)
knn

#Evaluate the model

predicted = knn.predict(x_test)
accuracy = accuracy_score(y_test, predicted)

print(f'The accuracy of the model is: {accuracy}')
```

### 13. Build a machine learning model to predict the sales of a product based on the advertising expenditure on TV, radio and newspaper. Use the 80:20 ratio to split the data. Evaluate the model using an appropriate evaluation metric. [15]

```python
advertising_df = pd.read_csv(r"C:\Users\ort_d\Downloads\advertising.csv")
advertising_df.head()
```

```python
#Select x and y features

x = advertising_df[['TV','Radio','Newspaper']]
y = advertising_df[['Sales']]

#Split the training and testing features
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

#Train the model
lm = LinearRegression()

lm.fit(x_train, y_train)

#Evaluate the model

predicted_val = lm.predict(x_test)

acc_score = r2_score(y_test, predicted_val)

print(acc_score)
```
