```python
import seaborn as sb

p_e_m = df.groupby('parental_level_of_education')['math_score'].mean().reset_index()
p_e_r = df.groupby('parental_level_of_education')['reading_score'].mean().reset_index()
p_e_w = df.groupby('parental_level_of_education')['writing_score'].mean().reset_index()

x = p_e_m['parental_level_of_education']
ym = p_e_m['math_score']
yr = p_e_r['reading_score']
yw = p_e_w['writing_score']

sb.lineplot(x = x, y = ym, label='math_score')
sb.lineplot(x = x, y = yr, label='reading_score')
sb.lineplot(x = x, y = yw, label='writing_score')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Average Scores')
plt.title('Average Scores by Parental Level of Education')
plt.legend()
plt.grid()
plt.show()
```
